```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the execution times from the CSV file
data = pd.read_csv("execution_times.csv")

# Extract algorithm names and their execution times
algorithms = data['Algorithm']
execution_times = data['Time']

# Create a bar chart
plt.figure(figsize=(10, 5))
plt.bar(algorithms, execution_times, color=['blue', 'orange', 'green'])

# Add labels and title
plt.xlabel('Algorithms')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of String Matching Algorithms Execution Time')
plt.ylim(0, max(execution_times) * 1.5)  # Set y-axis limit

# Display the execution time on top of each bar
for i, v in enumerate(execution_times):
    plt.text(i, v + 0.00002, f"{v:.6f}", ha='center', va='bottom')

# Show the plot
plt.grid(axis='y')
plt.show()
```
